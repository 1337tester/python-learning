```python
# modules we'll use
import pandas as pd
import numpy as np
import seaborn as sns
import datetime

# read in our data
landslides = pd.read_csv("catalog.csv")

# set seed for reproducibility
np.random.seed(0)

landslides.head()
```

```python
# print the first few rows of the date column
print(landslides['date'].head())
```

```python
# check the data type of our date column
landslides['date'].dtype
```

```python
# create a new column, date_parsed, with the parsed dates
landslides['date_parsed'] = pd.to_datetime(landslides['date'], format="%m/%d/%y")
```

```python
# print the first few rows
landslides['date_parsed'].head()
```

```python
# in case there are two different date formats in one column (it is slower though)
# landslides['date_parsed'] = pd.to_datetime(landslides['Date'], infer_datetime_format=True)
```

```python
# get the day of the month from the date_parsed column
day_of_month_landslides = landslides['date_parsed'].dt.day
day_of_month_landslides.head()
```

```python
# remove na's
day_of_month_landslides = day_of_month_landslides.dropna()

# plot the day of the month
sns.set_style("whitegrid")
sns.distplot(day_of_month_landslides, kde=False, bins=31)
```

```python
# excercise with earthquakes
earthquakes = pd.read_csv("earthquakes.csv")

# set seed for reproducibility
np.random.seed(0)

# earthquakes.head()
earthquakes["Date"].head()
```

```python
# different date format in line 3378
earthquakes[3378:3383]
```

```python
# check length of all dates
date_lengths = earthquakes.Date.str.len()
date_lengths.value_counts()
```

```python
# get those indices
indices = np.where([date_lengths == 24])[1]
print('Indices with corrupted data:', indices)
earthquakes.loc[indices]
```

```python
# 1975-02-23T02:58:41.000Z -> 3378  7512 20650 "%Y-%m-dT%H:%M:%S.%fZ"
# 02/23/1975 "%m/%d/%Y"
earthquakes.loc[3378, "Date"] = "02/23/1975"
earthquakes.loc[7512, "Date"] = "04/28/1985"
earthquakes.loc[20650, "Date"] = "03/13/2011"

earthquakes['date_parsed']  = pd.to_datetime(earthquakes['Date'], format="%m/%d/%Y")

earthquakes.head()
```

```python
day_of_month_earthquakes = earthquakes['date_parsed'].dt.day

day_of_month_earthquakes.head(5)
```

```python
# remove na's
day_of_month_earthquakes = day_of_month_earthquakes.dropna()

# plot the day of the month
sns.set_style("whitegrid")
sns.distplot(day_of_month_earthquakes, kde=False, bins=31)
```

```python
# volcano data past 10k years

volcanos = pd.read_csv("volcanic.csv")
volcanos.head()
```
